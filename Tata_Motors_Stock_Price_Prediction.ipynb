```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("TTM.csv")
print(data.head())
```

```python
# Candlestick charts --> used to represent the price movement of an asset. It provides a visual representation of
# the open, high, low, and close prices for each time period, typically displayed as rectangular "candles" on the chart.
#
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Tata Motors Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
print(data.corr())
```

AutoTS is an automated time series forecasting tool that utilizes various algorithms and techniques to automatically select and fit the best model for the given data.



```python
from autots import AutoTS
# creates an instance of the AutoTS class
# forecast_length=5 refers to predicting the values of a time series for the upcoming 5 periods or
# points in time, based on the historical data available.

# frequency='infer--> represents how often the data points are collected or measured. The frequency could be daily,
# monthly, quarterly, yearly, or any other interval depending on the nature of the data.


# ensemble='simple'--> combines the predictions of multiple models to generate an aggregated forecast for the given time series data.
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')

#  id_col=None--> indicating that there is no column representing individual IDs in the dataset.
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()   #  generates the predictions using the trained model.
forecast = prediction.forecast #  extracts the forecasted values from the prediction object and assigns them to the forecast variable.
print(forecast)
```
